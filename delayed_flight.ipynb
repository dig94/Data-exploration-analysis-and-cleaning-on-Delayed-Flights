```python
df2['DATE'] = pd.to_datetime(df2[['Year','Month','Day']])
```

```python
#function that convert the 'HHMM' string to datetime.time
import datetime
def format_time(time):
    if pd.isnull(time):
        return np.nan
    else:
        if time==2400:time=0
        time = "{0:04d}".format(int(time))
        time=datetime.time(int(time[0:2]),int(time[2:4]))
        return time
```

```python
df2['DepTime']=df2['DepTime'].apply(format_time)
df2['ArrTime']=df2['ArrTime'].apply(format_time)
df2['sheduledDepTime']=df2['sheduledDepTime'].apply(format_time)
df2['scheduledArrTime']=df2['scheduledArrTime'].apply(format_time)
df2.head(2)
```

```python
#extracting statistical parameters on each airline
def get_stats(group):
    return {'min': group.min(), 'max': group.max(),
            'count': group.count(), 'mean': group.mean()}
#_______________________________________________________________
# Creation of a dataframe with statitical infos on each airline:
global_stats = df3['DepDelay'].groupby(df3['UniqueCarrier']).apply(get_stats).unstack()
global_stats = global_stats.sort_values('count')
global_stats
```
